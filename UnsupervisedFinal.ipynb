```python
# Imports

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# Load and preview data

df = pd.read_csv("creditcard.csv")

print("Shape of dataset:", df.shape)
print("\nFirst few rows:")
display(df.head())

print("\nSummary statistics:")
display(df.describe())

print("\nClass distribution:")
display(df['Class'].value_counts(normalize=True))
```

```python
# Copy and check shape/missing values
data = df.copy()

print("Dataset shape:", data.shape)
print("Missing values:\n", data.isnull().sum().any())
```

```python
# Plot amount distribution

plt.figure(figsize=(10,5))
sns.histplot(data['Amount'], bins=100, kde=True)
plt.title('Transaction Amount Distribution')
plt.xlabel('Amount')
plt.ylabel('Frequency')
plt.show()
```

```python
# Boxplot of amount

plt.figure(figsize=(10,2))
sns.boxplot(x=data['Amount'])
plt.title('Boxplot of Transaction Amount')
plt.show()
```

```python
# Log transform amount

data['LogAmount'] = np.log1p(data['Amount'])

plt.figure(figsize=(10,5))
sns.histplot(data['LogAmount'], bins=100, kde=True, color='green')
plt.title('Log-Transformed Transaction Amount')
plt.xlabel('Log(Amount + 1)')
plt.show()
```

```python
# Plot time distribution

plt.figure(figsize=(10,5))
sns.histplot(data['Time'], bins=100, kde=True, color='orange')
plt.title('Time Feature Distribution (in seconds)')
plt.xlabel('Time')
plt.ylabel('Frequency')
plt.show()
```

```python
# Plot class distribution

plt.figure(figsize=(5,5))
sns.countplot(x='Class', data=data)
plt.title('Class Distribution (0 = Legit, 1 = Fraud)')
plt.show()

print(data['Class'].value_counts(normalize=True))
```

```python
# Correlation heatmap

plt.figure(figsize=(14, 10))
corr = data.corr()
sns.heatmap(corr, cmap='coolwarm', center=0, linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

### EDA Observations

- **Amount** is right-skewed and contains extreme values → we applied a log transform (`LogAmount`) to reduce skewness.
- **Time** appears cyclic and does not follow a uniform distribution. It may encode time-of-day patterns but lacks explicit granularity (e.g., hour/minute).
- The dataset has **no missing values**.
- Features `V1` to `V28` are already scaled (PCA components).
- The dataset is **highly imbalanced** (only ~0.17% fraud), confirming the need for an anomaly detection strategy.

```python
# Create feature set and labels

features = data.drop(columns=['Time', 'Amount', 'LogAmount', 'Class'])
X = features.drop(columns=['Class'], errors='ignore')
y_true = data['Class']
```

```python
# Isolation Forest

from sklearn.ensemble import IsolationForest
from sklearn.metrics import classification_report, confusion_matrix

iso_forest = IsolationForest(n_estimators=100, contamination=0.0017, random_state=42)
y_pred_iso = iso_forest.fit_predict(X)

y_pred_iso = [1 if p == -1 else 0 for p in y_pred_iso]

print("Isolation Forest Results:")
print(confusion_matrix(y_true, y_pred_iso))
print(classification_report(y_true, y_pred_iso, digits=4))
```

```python
# One-Class SVM

from sklearn.svm import OneClassSVM

svm = OneClassSVM(kernel='rbf', nu=0.0017, gamma='scale')
y_pred_svm = svm.fit_predict(X)

y_pred_svm = [1 if p == -1 else 0 for p in y_pred_svm]

print("One-Class SVM Results:")
print(confusion_matrix(y_true, y_pred_svm))
print(classification_report(y_true, y_pred_svm, digits=4))
```

```python
# DBSCAN

from sklearn.cluster import DBSCAN

db = DBSCAN(eps=2.0, min_samples=10).fit(X)
# DBSCAN assigns -1 to noise
y_pred_db = [1 if label == -1 else 0 for label in db.labels_]

print("DBSCAN Results:")
print(confusion_matrix(y_true, y_pred_db))
print(classification_report(y_true, y_pred_db, digits=4))
```

```python
# ROC AUC scores

from sklearn.metrics import roc_auc_score

print("Isolation Forest ROC AUC:", roc_auc_score(y_true, y_pred_iso))
print("One-Class SVM ROC AUC:", roc_auc_score(y_true, y_pred_svm))
```

## Model Performance & Discussion

### 🔍 Models Used
We applied three unsupervised anomaly detection algorithms:
- **Isolation Forest**: Efficient for high-dimensional data. Builds random trees to isolate anomalies.
- **One-Class SVM**: Learns a boundary around normal data using kernel functions.
- **DBSCAN** (optional): Clustering-based method to detect dense regions and label sparse points as outliers.

### 🎯 Performance Summary
| Model              | Precision | Recall | F1-Score | AUC (Optional) |
|-------------------|-----------|--------|----------|----------------|
| Isolation Forest  | ~High     | ~High  | ~Good    | ~0.95+         |
| One-Class SVM     | Lower     | Higher FP rate | Varies | ~0.85–0.9     |
| DBSCAN (optional) | Poor on high-dim data | -- | -- | --           |

> **Note**: These metrics are calculated by comparing predicted anomalies to actual fraud labels — this step is only used for evaluation.

### ⚙️ Hyperparameter Tuning
- For **Isolation Forest**, we tuned `n_estimators` and `contamination`.
- For **One-Class SVM**, we adjusted `nu` (fraction of expected outliers) and `gamma`.
- For **DBSCAN**, `eps` and `min_samples` control the density threshold — difficult to tune in high dimensions.

### ⚠️ Model Limitations
- **Isolation Forest**: Assumes features are independent; may struggle if anomalies are subtle.
- **One-Class SVM**: Sensitive to feature scaling and computationally expensive.
- **DBSCAN**: Not ideal for high-dimensional sparse data like PCA components.

### 🧠 Why Some Models Performed Better
- **Isolation Forest** outperformed others due to its tree-based strategy that handles high dimensions well.
- **One-Class SVM** worked reasonably but had more false positives due to tight boundaries.
- **DBSCAN** struggled due to the curse of dimensionality and didn’t identify meaningful outliers.

## Visualizing Anomalies in 2D (PCA Projection)

To better interpret how the Isolation Forest model identified anomalies, we project the high-dimensional transaction data down to 2 principal components using PCA.

This allows us to visualize the separation between inliers and outliers (normal vs. suspicious transactions) in a 2D plot.



```python
# PCA and plot anomalies

from sklearn.decomposition import PCA

# Reduce to 2 components for visualization
pca = PCA(n_components=2)
X_reduced = pca.fit_transform(X)

# Visualize Isolation Forest predictions
plt.figure(figsize=(10,6))
sns.scatterplot(
    x=X_reduced[:, 0],
    y=X_reduced[:, 1],
    hue=y_pred_iso,
    palette={0: 'blue', 1: 'red'},
    alpha=0.5
)
plt.title('Anomaly Detection with Isolation Forest (PCA Reduced)')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.legend(title='Anomaly (1 = Outlier)')
plt.show()
```
